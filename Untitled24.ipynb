```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("creditcard1.csv")
```

```python
data.head()
```

```python
data.tail()
#last elements
```

```python
data.isnull().sum()
```

```python
data['Class'].value_counts()
#Finding no of fraud transcation
```

```python
legit = data[data.Class==0]#separating legit and fraud transcations
fraud=data[data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
#statistical information of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
data.groupby("Class").mean()
#comparing both transcations side by side
```

```python
legit_sample=legit.sample(n=492)
#concatenating 2 data frames
```

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby("Class").mean()
```

```python
X=new_df.drop(columns="Class",axis=1)
Y=new_df['Class']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)#model training
```

```python
#FOR BINARY CLASSIFICATION
model=LogisticRegression()
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
#accuracy score on training data
```

```python
print("the accuracy of training data:",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
#accuracy score on testing data
```

```python
print("the accuracy of testing data:",test_data_accuracy)
```
